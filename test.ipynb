```python
import csv
from typing import List
import os

def get_csv_chapters(path: str, start: int, end: int):
    with open(path, mode='r', encoding='utf-8') as f:
        reader = csv.DictReader(f)
        headers = reader.fieldnames
        print(f"CSV Headers: {headers}")
        chapters = []
        for row in reader:
            chapters.append((row['Chapter'], row['Content']))

        num_chapters = len(chapters)
        print(f'Total number of chapters: {num_chapters}')

        text = ""
        if start + end > num_chapters:
            print(f"This file only has {num_chapters} chapters. Using all of them.")
            for _, content in chapters:
                text += content + "\n"
        else:
            # Get the first 'start' chapters
            for _, content in chapters[:start]:
                text += content + "\n"
            # Get the last 'end' chapters
            for _, content in chapters[-end:]:
                text += content + "\n"
    
    return text
```

```python
get_csv_chapters("data/ac_csv/poirot_investigates.csv", 2, 1)
```
